```python
# drop month and year on imported .csv
pce = pce.drop(['Month', 'Year'], axis=1)
```

```python
# create fiscal year offset variable
fiscal_month_offset = 6
```

```python
# make a copy of the df
pce_1=pce.copy()
```

```python
# split fiscal year from dimension code and add to 2000
pce_1['Fiscal_Year']= 2000 + np.floor_divide(pce_1['Shortcut_Dimension_4_Code'],1000000)
```

```python
# Create Fiscal Date column 
pce_1['Fiscal_Date'] = pce_1.apply(lambda row: date(row['Fiscal_Year'], row['Stage_Code'], 1), axis=1)
```

```python
# subtract fiscal date from fiscal offset and create a new column
pce_1['Calendar_Date'] = pce_1['Fiscal_Date'] - relativedelta(months=fiscal_month_offset)
pce_1['Calendar_Date']= pd.to_datetime(pce_1['Calendar_Date'])
```

```python
# strip month and year from datetime and convert to a string
pce_1['Month']= pce_1['Calendar_Date'].dt.strftime('%m')
pce_1['Year']= pce_1['Calendar_Date'].dt.strftime('%Y')
```

```python
# convert Month and Year String to integer
pce_1['Month'] = pce_1.Month.astype(int)
pce_1['Year'] = pce_1.Year.astype(int)
```

```python
pce=pce_1.copy()
```

```python
pce.head()
```

```python
#Sum amount from line no 10000, 20000, 30000, 40000 into 10000 and creat new column
grouped_pce_amount=pce.groupby(['Community_Name','Month','Year'])[["Amount"]].sum()
grouped_pce_amount = grouped_pce_amount.reset_index()
grouped_pce_amount = grouped_pce_amount.rename(columns={'Amount':'Amount_Combined'})
pce_2 = pd.merge(pce, grouped_pce_amount, how='left', 
                          left_on=['Community_Name','Month', 'Year'], right_on=['Community_Name', 'Month', 'Year'])
```

```python
pce_2.shape
```

```python
#remove line no 20000, 30000, 40000 values as they are redundant now
pce_3=pce_2.drop(pce_2[pce_2.Line_No == 20000].index)
pce_4=pce_3.drop(pce_3[pce_3.Line_No == 30000].index)
pce_5=pce_4.drop(pce_4[pce_4.Line_No == 40000].index)

pce=pce_5.copy()
pce.shape
```

```python
#Calculations for line loss, new line loss columns added to df
pce['Total_Sold']= pce.Residential_Sold_To + pce.Commercial_Sold_To + pce.Com_Facil_Sold_To + pce.Govt_Facil_Sold_To + pce.Unbilled_Sold_To
pce['Total_Generated']= pce.Diesel_kWh_Generated + pce.Hydro_kWh_Generated + pce.Other_1_kWh_Generated  + pce.Other_2_kWh_Generated + pce.Total_kWh_Purchased
pce['Total_Adjusted_Generation']= pce.Total_Generated - pce.Powerhouse_Consumption_kWh

pce['Line_Efficiency']= round(pce.Total_Sold / pce.Total_Adjusted_Generation,4)
pce['Line_Loss']= 1- pce.Line_Efficiency

line_loss = ['Residential_Rate','Year_Intertied','Community_Name','Calendar_Date', 'Total_Sold', 'Total_Generated', 'Total_Adjusted_Generation', 
             'Line_Loss', 'receiving_intertied_power']

```
